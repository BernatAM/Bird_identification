```python
# Leemos el texto
text = open('aves.txt', 'rb')
content = text.read()
print(content)
```

```python
# Pasos que vamos a seguir
# quitamos todas las palabras que no queremos (nos quedamos con verbos, nombres y adjetivos)
# tokenizamos todo el texto
# ahora consultamos la dbpedia y obtenemos todos los pajaros
# tockenizamos los pajaros
# lso buscamos dentro del texto encontrarlos
```

```python
# quitamos todas las palabras que no queremos (nos quedamos con verbos, nombres y adjetivos)
import nltk
from nltk.stem import PorterStemmer

def prepara_texto(content):
    frases = nltk.sent_tokenize(str(content))
    texto_tokenizado_filtrado = []
    Texto_token_filt = ' '

    stemmer = PorterStemmer()
    for frase in frases:
        Frases_gramatical = nltk.pos_tag(frase.split(), tagset='universal')

    # tokenizamos todo el texto  
        for palabra,categoria in Frases_gramatical:
            if categoria in ['NOUN','VERB', 'ADJ']:
                raiz = stemmer.stem(palabra)
                texto_tokenizado_filtrado.append(raiz)
                Texto_token_filt = Texto_token_filt +' ' +raiz
    return (Texto_token_filt)            
texto_preparado=prepara_texto(content)
```

```python
### Ahora tenemos que consultar en la dbpedia los pajaros que estamos buscando para poder comparar
query = """ prefix dbo:<http://dbpedia.org/ontology/> select * where {
    ?x a dbo:Bird .
    ?x rdfs:label ?label .
    FILTER (lang(?label) = 'en')
    
    } """

# con esta query sacamos todos los tipos de pajaros
```

```python
#hacemos la query a la dbpedia

from SPARQLWrapper import SPARQLWrapper2

sparql = SPARQLWrapper2("http://dbpedia.org/sparql")
sparql.setQuery(query)

Pajaros = []
enlaces = []
for result in sparql.query().bindings:
    #print(f"{result['label'].lang}, {result['label'].value}")
    Pajaros.append(f" {result['label'].value}")
    enlaces.append(f" {result['x'].value}")
    
print (Pajaros)
print (enlaces)
```

```python
# juntamos el nombre del pajaro y del enlace en la dbpedia
Pajaros_enlace=tuple(zip(Pajaros, enlaces))
```

```python
# el siguiente paso es tokenizar los diferentes tipos de pajaros para que tengan la misma forma que en el texto
Pajaros_tokenizados = []
stemmer = PorterStemmer()
for pajaro in Pajaros:
    pajaro_tokenizado = ''
    splited_bird =pajaro.split()
    for palabra in splited_bird: 
        pajaro_tokenizado = " ".join([pajaro_tokenizado, stemmer.stem(palabra)])
    Pajaros_tokenizados.append(pajaro_tokenizado)
    
print ( Pajaros_tokenizados)    
    
```

```python
# juntamos los pajaros tokenizados con los pajaros sin tokenizar y lo enlaces en forma de diccionario

Pajaros_dict = {i:[j, k] for i, j, k in zip(Pajaros, Pajaros_tokenizados, enlaces)}
print (Pajaros_dict)
```

```python
# una vez modificados los pajaros podemos tratar de identificarlos en el texto


def find_birds(text):
    bird_found = []
    for pajaro in Pajaros:
        if Pajaros_dict[pajaro][0] in text:
            print(Pajaros_dict[pajaro][1])
            bird_found.append(pajaro)
    return (bird_found)
find_birds(texto_preparado)
```

```python
# ya tenemos la función que nos encuentra los pajaros en un texto
# definimos la estructura de las ontologias

from owlready2 import *

onto = get_ontology("http://test.org/onto_final.owl")


with onto:
    class bird(Thing):
        pass
    class file(Thing):
        pass 
    class dbpedia_bird(Thing):
        pass
    
    class url(ObjectProperty):
        domain = [bird]
        range = [dbpedia_bird]
    class find_in(ObjectProperty):
        domain = [bird]
        range = [file]
```

```python
def create_ontology_from_file(file):
    text = open('files/'+file, 'rb')
    content = text.read()
    texto_preparado = prepara_texto(content)
    pajaros_encontrados=find_birds(texto_preparado)
    for pajaro in pajaros_encontrados:
        print (pajaro)
        my_pajaro = bird(pajaro.replace(' ', '_'))
        my_file = file_from(str(file))
        my_pajaro.find_in = [my_file]
        my_dbird =dbpedia_bird(str(Pajaros_dict[pajaro][1]))
        my_pajaro.url = [my_dbird]
    onto.save(file="Trabajo_final_ontology.xml", format = "rdfxml")
    
    


```
